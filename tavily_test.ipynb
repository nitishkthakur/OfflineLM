```python
from tavily import TavilyClient
import asyncio
from IPython.display import display, Markdown
tavily_client = TavilyClient()
```

```python
# Step 1: Search with advanced parameters
search_response = tavily_client.search(
    query="Who is the current president of the US?",
    search_depth="advanced",
    include_raw_content=True,
    max_results=10
)

# Step 2: Filter high-quality URLs
relevant_urls = [
    result['url'] for result in search_response['results'] 
    if result.get('score', 0) > 0.5
]

# Step 3: Extract full content with advanced depth
extracted_content = await asyncio.gather(
    *[tavily_client.extract(url, extract_depth="advanced") for url in relevant_urls]
)
```

```python
def tavily_to_context(search_results):
    context = []
    for result in search_results['results']:
        context.append(result['raw_content'])

    return "\n\n\n".join(context)


Markdown(tavily_to_context(search_response))
```
